```python
crop_data = pd.read_csv("dataset.csv")
```

```python
crop_data.isnull().sum()
```

```python
crop_data = crop_data.dropna()
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelEncoder()
crop_data['State'] = lb.fit_transform(crop_data['State'])
crop_data['Crop'] = lb.fit_transform(crop_data['Crop'])
print(crop_data)
```

```python
from sklearn.model_selection import train_test_split

x = crop_data.drop(["Yield","Cost"], axis=1)
y = crop_data["Yield"]
print(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=5)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
import joblib
joblib.dump(model,'mymodel_for_cropprice.pkl')
```
